```python
LS_candidate_combined["Coalition"]=LS_candidate_combined["Party Abbreviation"]
parties_UPA = ['INC','NCP', 'RJD', 'DMK', 'IUML', 'JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD']
for i in range(len(parties_UPA)):
    LS_candidate_combined["Coalition"] = LS_candidate_combined["Coalition"].str.replace(parties_UPA[i],'UPA')

```

```python
party_NDA = ['BJP','SS', 'LJP', 'SAD', 'RLSP', 'AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS','AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']

for i in range(len(party_NDA)):
    LS_candidate_combined["Coalition"] = LS_candidate_combined["Coalition"].str.replace(party_NDA[i],'NDA')
    
```

```python
  LS_candidate_combined
```

```python
party_others = ['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM', 'KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL', 'JVM','MDMK']
for i in range(len(party_others)):
        LS_candidate_combined["Coalition"] = LS_candidate_combined["Coalition"].str.replace(party_others[i],'Others')

    
```

```python
LS_candidate_combined["Coalition"].value_counts()
```

```python
Filter_Candidates =LS_candidate_combined["Position"] == 1
```

```python
LS_party_Seats_Won=LS_candidate_combined[Filter_Candidates].groupby(["Coalition","Year"])["Position"].sum().reset_index().pivot(index = "Coalition", values = "Position", columns = "Year").fillna(0).reset_index()
```

```python
LS_party_Seats_Won=(pd.DataFrame(LS_party_Seats_Won))
```

```python
LS_party_Seats_Won
```

```python
LS_party_Seats_Won = LS_party_Seats_Won.sort_values(by=[2009,2014], ascending= False)
```

```python
colors = ["orange", "green", "red", "cyan", "brown", "grey", "blue", "indigo", "beige", "yellow","cadetblue","khaki"]
sizes = LS_party_Seats_Won[2009]
labels = LS_party_Seats_Won["Coalition"]
explodes = (0.05,0.05,0.05,0.05)
plt.figure(figsize =(20,8))
plt.subplot(1,2,1)

plt.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%')

draw_circle = plt.Circle((0,0), 0.70, fc = "white")
#for existing figure 
fig = plt.gcf() # means get current figure
fig.suptitle("Winning Percentages by Alliances and Major Political Parties", fontsize=14) # Adding supertitle 
ax = fig.gca()  # get current axis
ax.annotate("2009", xy=(0, 0), fontsize=30, ha="center",va="center")
ax.add_artist(draw_circle)
#Equal aspect ratio ensures that pie is drawn as a circle
ax.axis("off")
ax.set_aspect('equal')

colors = ["orange", "green", "red", "cyan", "brown", "grey", "blue", "indigo", "beige", "yellow","cadetblue","khaki"]
sizes_2014 = LS_party_Seats_Won[2014]
labels_2014 = LS_party_Seats_Won["Coalition"]
explodes = (0.05,0.05,0.05,0.05)
plt.figure(figsize =(20,8))
plt.subplot(1,2,1)

plt.pie(sizes_2014, colors = colors, labels=labels_2014, autopct='%1.1f%%')

draw_circle2 = plt.Circle((0,0), 0.7, fc = "white")
#for existing figure 
fig1 = plt.gcf() # means get current figure
ax1 = fig1.gca()  # get current axis
ax1.annotate("2014", xy=(0, 0), fontsize=30, ha="center",va="center")
ax1.add_artist(draw_circle2)
#Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis("off")
ax1.set_aspect('equal')
plt.show()
```

```python
year_filter_2009 = LS_candidate_combined["Year"] == (2009)
position_filter_2009 = LS_candidate_combined["Position"] == 1.0

LS_candidate_combined[year_filter_2009]
```

```python
LS_new_Bar= LS_candidate_combined[year_filter_2009 & position_filter_2009].groupby(["Coalition","Candidate Category"])["Position"].sum().reset_index()
```

```python
LS_filter_2014 = LS_candidate_combined["Year"] == 2014
position_filter_2014 = LS_candidate_combined["Position"] == 1
```

```python
LS_new_2014_Bar = LS_candidate_combined[LS_filter_2014 & position_filter_2014].groupby(["Coalition", "Candidate Category"])["Position"].sum().unstack().reset_index()
LS_new_2014_Bar= LS_new_2014_Bar.fillna(0)
```

```python

# The bar plot has legent, you can specify the title for the graph, as well as one can set the legend to true. The bar plot helps us to decipher 
graph = LS_new_2014_Bar.plot(kind= "bar", title = "Winning Category for 2014", legend = True, fontsize = 12, figsize=(12,12))


graph.set_ylabel("Seats Won")
graph.set_xlabel("Parties")


labels = [i.get_text() for i in graph.get_xticklabels()]

for i in range(len(labels)):
    labels[i]  = LS_new_2014_Bar["Coalition"]
    graph.set_xticklabels(labels[i])




```
